```python
import pandas as pd
import re
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
Final_df = pd.read_csv('Final_features.csv')
```

```python
Final_df.head()
```

```python
SA = Final_df[Final_df.TARGETS.str.contains('Staphylococcus aureus')==True]
EC = Final_df[Final_df.TARGETS.str.contains('Escherichia coli')==True]
PA = Final_df[Final_df.TARGETS.str.contains('Pseudomonas aeruginosa')==True]
```

```python
#SA
SA_X_train, SA_X_test, SA_y_train, SA_y_test = train_test_split(SA, SA['NEW-CONCENTRATION'], test_size=0.2, random_state=30)
SA_X_train, SA_X_val, SA_y_train, SA_y_val = train_test_split(SA_X_train, SA_y_train, test_size=0.2, random_state=30)

#EC
EC_X_train, EC_X_test, EC_y_train, EC_y_test = train_test_split(EC, EC['NEW-CONCENTRATION'], test_size=0.2, random_state=30)
EC_X_train, EC_X_val, EC_y_train, EC_y_val = train_test_split(EC_X_train, EC_y_train, test_size=0.2, random_state=30)

#PA
PA_X_train, PA_X_test, PA_y_train, PA_y_test = train_test_split(PA, PA['NEW-CONCENTRATION'], test_size=0.2, random_state=30)
PA_X_train, PA_X_val, PA_y_train, PA_y_val = train_test_split(PA_X_train, PA_y_train, test_size=0.2, random_state=30)
```

```python
SA_X_train.to_csv('SA_X_train.csv',index =False)
SA_X_test.to_csv('SA_X_test.csv',index =False)
SA_X_val.to_csv('SA_X_val.csv',index =False)

EC_X_train.to_csv('EC_X_train.csv',index =False)
EC_X_test.to_csv('EC_X_test.csv',index =False)
EC_X_val.to_csv('EC_X_val.csv',index =False)

PA_X_train.to_csv('PA_X_train.csv',index =False)
PA_X_test.to_csv('PA_X_test.csv',index =False)
PA_X_val.to_csv('PA_X_val.csv',index =False)
```
